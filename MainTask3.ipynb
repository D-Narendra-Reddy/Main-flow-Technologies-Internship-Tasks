```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [30500, 35600, 28300, 33900, 39900, 41400, 48800, 53000, 48700, 43300, 37200, 45000],
    'Expenses': [22000, 23400, 20000, 19000, 23000, 24000, 24500, 26000, 28000, 29000, 32000, 31000]
}

df = pd.DataFrame(data)

# Set the Month column as the index
df.set_index('Month', inplace=True)

# Display the dataframe
print(df)
```

```python
# Bar chart for Sales
plt.figure(figsize=(10, 6))
df['Sales'].plot(kind='bar', color='skyblue', edgecolor='black')

# Customize the bar chart
plt.title('Monthly Sales', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales ($)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(['Sales'], loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the bar chart
plt.tight_layout()
plt.show()
```

```python
# Line chart for Sales and Expenses
plt.figure(figsize=(10, 6))
df['Sales'].plot(kind='line', marker='o', linestyle='-', color='b', label='Sales')
df['Expenses'].plot(kind='line', marker='x', linestyle='--', color='r', label='Expenses')

# Customize the line chart
plt.title('Monthly Sales and Expenses', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Amount ($)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.grid(axis='both', linestyle='--', alpha=0.7)

# Display the line chart
plt.tight_layout()
plt.show()
```

```python
# Histogram for Sales
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=10, color='skyblue', edgecolor='black')

# Customize the histogram
plt.title('Distribution of Monthly Sales', fontsize=16)
plt.xlabel('Sales ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the histogram
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for Sales vs. Expenses
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Expenses'], color='skyblue', edgecolor='black')

# Customize the scatter plot
plt.title('Sales vs. Expenses', fontsize=16)
plt.xlabel('Sales ($)', fontsize=14)
plt.ylabel('Expenses ($)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

# Display the scatter plot
plt.tight_layout()
plt.show()
```
